# **Pre-Trained Object Detection Model**

## **YOLO based object detection**

YOLO (You Only Look Once) is a cutting-edge object detection framework designed to detect and localize multiple objects in a single glance. Unlike older approaches that scan an image in multiple passes, YOLO looks at the entire image just once and predicts all bounding boxes and class probabilities simultaneously. This unified architecture makes it blazing fast and highly efficient, which is why itâ€™s widely used in real-time applications like autonomous driving, video surveillance, and industrial automation.

The model comes in various sizes (like yolo11n.pt, the nano version), which trade off accuracy for speed. The nano variant is exceptionally lightweight, making it ideal for quick prototyping or running on resource-limited hardware.

## **Step 3 - Load Pre-trained YOLOv11 model**

Ultralytics offers multiple model sizes for each YOLO version, each trading accuracy for speed.
You can plug any of these `.pt` weights into the same `YOLO()` call:

| Model        | Description                                             |
| ------------ | ------------------------------------------------------- |
| `yolo11n.pt` | **Nano** â€” ultra-light, fastest, good for edge devices. |
| `yolo11s.pt` | **Small** â€” slightly larger, better accuracy than nano. |
| `yolo11m.pt` | **Medium** â€” balanced speed and accuracy.               |
| `yolo11l.pt` | **Large** â€” higher accuracy, slower inference.          |
| `yolo11x.pt` | **Extra-large** â€” best accuracy, heavy compute load.    |

Here we use `yolo11n.pt`,


```python
model = YOLO("yolo11n.pt")
```

## **Step 4 â€” Test with a Single Image**
We run the image through the model, and get bounding boxes + labels drawn automatically.
* conf=0.25 - filters out low-confidence predictions to reduce noise.
* .show() - opens a window with the detections overlaid.

```python
img_path = "/content/sample.jpg"
results = model(img_path, conf=0.25)
results[0].show()
```
